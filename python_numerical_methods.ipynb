# Вопрос 1
### Указать порядок абсолютной погрешности числа:

```python
import re

def solve(str_num):
    decimal = re.search(r'(?:\.\d*?)(\d+)', str_num.replace(',', '.'))
    return -len(decimal.group(1)) if decimal else 0

x1 = '555.07888'    # ЗАМЕНИТЬ НА СВОЕ   
x2 = '10.0700000'   # ЗАМЕНИТЬ НА СВОЕ

print(f"Ответ: {solve(x1)} ; {solve(x2)}")
```

# Вопрос 2
### Числа X<sub>1</sub> и X<sub>2</sub> содержат N верных цифр в широков смысле. Определить их абсолютную погрешность


```python
def solve(str_num, n):
    parts = str_num.replace('.', ',').split(',')
    if len(parts) == 2:
        if len(parts[0]) < n:
            return '0,' + '0' * (n - len(parts[0]) - 1) + '1'
        else:
            return '0,1'
    else:
        return '0,' + '0' * (n - 2) + '1'

x1 = '2.99285673'       # ЗАМЕНИТЬ НА СВОЕ
x2 = '48628.27453'      # ЗАМЕНИТЬ НА СВОЕ
n = 7                   # N ЗАМЕНИТЬ НА СВОЕ (БЕЗ КАВЫЧЕК)

print(f"Ответ: {solve(x1, n)} ; {solve(x2, n)}")
```

# Вопрос 3
### Абсолютная погрешность чисел X<sub>1</sub> и X<sub>2</sub> составляет A. Округлить числа, оставив только верные цифры

```python
def solve(num_str, error_str):
    num = float(num_str.replace(',', '.'))
    error = float(error_str.replace(',', '.'))
    multiplier = 1
    while error * 10 < 1:
        multiplier *= 10
        error *= 10
    rounded_num = round(num * multiplier) / multiplier
    if rounded_num % 1 == 0:
        return str(int(rounded_num)).replace('.', ',')
    else:
        return str(round(rounded_num, 10)).replace('.', ',')

x1 = '9.91565'    # ЗАМЕНИТЬ НА СВОЕ
x2 = '694.75231'  # ЗАМЕНИТЬ НА СВОЕ
a = '0.004'       # A ЗАМЕНИТЬ НА СВОЕ

print(f"Ответ: {solve(x1, a)} ; {solve(x2, a)}")

```

# Вопрос 4
### Найти мантиссу и порядок числа X

```python
import math

def solve(str_num):
    number = float(str_num.replace(',', '.'))
    if number == 0:
        mantissa = 0
        exponent = 0
    else:
        exponent = int(math.log10(abs(number)))
        mantissa = number / (10 ** exponent)
    
    return mantissa, exponent

number = '730'  # ЗАМЕНИТЬ НА СВОЕ

ans = solve(number)
print(f"Ответ: {str(ans[0]).replace('.', ',')} ; {ans[1]} (Не засчитывает как правильный ответ)")
```

# Вопрос 5
### Вычислить a=, где все цифры чисел верные. Через точку с запятой указать допускаемую абсолютную погрешность

```python
import re

def derivative(x):
    return '0.' + '0' * (len(x.split('.')[1]) - 1) + '1'

def solve(expression):
    expression = expression.replace(" ", "").replace(",", ".")
    floats = re.findall(r'\d+\.\d+', expression)
    operations = re.findall(r'[+\-]', expression)
    derivative_floats = [float(derivative(i)) for i in floats]
    decimal_places = [len(f.split('.')[1]) for f in floats]
    less = min(decimal_places)
    rounded_floats = [round(float(f), less + 1) for f in floats]
    rounded_floats = [str(num) for num in rounded_floats]
    res = ''.join(
        [f"{operations[idx - 1]}{num}" if idx != 0 else num for idx, num in enumerate(rounded_floats)]
    )
    result = round(eval(res), less)
    return result, round(sum(derivative_floats), max(decimal_places))

expression = "719.05926+5.046-0.00961"  # ЗАМЕНИТЬ НА СВОЕ

ans, error = solve(expression)
print(f"Ответ: {str(ans).replace('.', ',')} ; {str(error).replace('.', ',')}")
```

# Вопрос 6
### Найти порядки предельных относительных погрешностей чисел X<sub>1</sub> и X<sub>2</sub>, если все цифры верные

```python
import re

def parse(x):
    x = x.replace(',', '.')
    int_part, float_part = x.split('.')
    int_part = re.sub(r'^\d*?0+', '', int_part)
    
    if int_part:
        return [int(int_part[0]), len(int_part) - 1, -len(float_part)]
    else:
        for i, d in enumerate(float_part):
            if d != '0':
                return [int(d), -1 - i, -len(float_part)]
        return [0, -len(float_part), -len(float_part)]

def solve(x):
    dat = parse(x)
    return dat[2] - dat[1]

x1 = '69.6740'   # ЗАМЕНИТЬ НА СВОЕ
x2 = '0.001798'  # ЗАМЕНИТЬ НА СВОЕ

print(f"Ответ: {solve(x1)} ; {solve(x2)}")
```

# Вопрос 7
### Используя правила, найти произведение приближенных чисел X<sub>1</sub> и X<sub>2</sub>, все цифры которых верные. В ответе указать полученное произведение и, через точку с запятой, порядок его предельной относительной погрешности

```python
import re

def parse(x):
    int_part, float_part = x.split('.')
    int_part = re.sub(r'^\d*?0+', '', int_part)
    if int_part:
        return [int(int_part[0]), len(int_part) - 1, -len(float_part)]
    else:
        for i, d in enumerate(float_part):
            if d != '0':
                return [int(d), -1 - i, -len(float_part)]
        return [0, -len(float_part), -len(float_part)]

def solve(x1, x2):
    x1 = float(x1.replace(',', '.'))
    x2 = float(x2.replace(',', '.'))
    less = min(len(str(x1).split('.')[-1]), len(str(x2).split('.')[-1]))
    if len(str(x1).split('.')[-1]) > less:
        x1 = round(x1, less + 1)
    if len(str(x2).split('.')[-1]) > less:
        x2 = round(x2, less + 1)
    res = f"{round(x1 * x2, less):.{less}f}"
    order = parse(res)
    return res.replace('.', ','), order[2] - order[1]

x1 = '4.86'     # ЗАМЕНИТЬ НА СВОЕ
x2 = '0.84888'  # ЗАМЕНИТЬ НА СВОЕ

product, order = solve(x1, x2)
print(f"Ответ: {product} ; {order}")
```
